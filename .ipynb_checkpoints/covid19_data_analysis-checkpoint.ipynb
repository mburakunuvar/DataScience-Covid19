## Prepare Data

```python
#worldwide-aggregated data with date as column
df_ww2 = pd.read_csv('./data/worldwide-aggregated_csv.csv', parse_dates=True)

# convert data type from object to datetime
# https://www.geeksforgeeks.org/convert-the-column-type-from-string-to-datetime-format-in-pandas-dataframe/
df_ww2['Date']=pd.to_datetime(df_ww2['Date'])

# filter data as of 1st of February 
df_ww2 = df_ww2.query('Date >= "2020-02-01"')

# generate name of month of from date
# https://stackoverflow.com/questions/57033657/how-to-extract-month-name-and-year-from-date-column-of-dataframe/57033774
df_ww2['Month'] = df_ww2['Date'].dt.strftime('%b')
```

```python
df_ww2.head()
```

```python
df_ww2.rename(columns={'Confirmed': 'Confirmed_cumulative', 'Recovered': 'Recovered_cumulative','Deaths': 'Deaths_cumulative'}, inplace=True)
```

```python
df_ww2.head()
```

```python
def display_barplot(df, columns):
    '''
    INPUT - df - pandas dataframe to be analysed & visualized
            columns - array- the name 3 features to be analyzed : confirmed, recovered, deaths
            
    OUTPUT -
            graph - multiple bar plot visualizing the Covid19 Numbers during each month 
    '''
    for item in columns:
        df[item+'_daily'] = df[item+'_cumulative']-df[item+'_cumulative'].shift(1)
    confirmed = df.groupby('Month').Confirmed_daily.sum().astype(int)
    recovered = df.groupby('Month').Recovered_daily.sum().astype(int)
    deaths = df.groupby('Month').Deaths_daily.sum().astype(int)
    new_order = ['Feb', 'Mar', 'Apr', 'May', 'Jun']

    # set width of bar 
    barWidth = 0.3
    fig = plt.subplots(figsize =(9, 7))
    
    # set height of bar 
    confirmed= confirmed.reindex(new_order, axis=0).to_list()
    recovered= recovered.reindex(new_order, axis=0).to_list()
    deaths= deaths.reindex(new_order, axis=0).to_list()
    
    # Set position of bar on X axis 
    br1 = np.arange(len(confirmed)) 
    br2 = [x + barWidth for x in br1] 
    br3 = [x + barWidth for x in br2] 
    
    # Make the plot 
    # https://matplotlib.org/3.1.1/gallery/color/named_colors.html
    plt.bar(br1, confirmed, color ='gold', width = barWidth, 
            edgecolor ='grey', label ='Confirmed') 
    plt.bar(br2, recovered, color ='darkolivegreen', width = barWidth, 
            edgecolor ='grey', label ='Recovered') 
    plt.bar(br3, deaths, color ='r', width = barWidth, 
            edgecolor ='grey', label ='Deaths') 
    
    # Adding Xticks  
    plt.xlabel('Months - 2020', fontweight ='bold') 
    plt.ylabel('Reported Numbers', fontweight ='bold') 
    plt.xticks([r + barWidth for r in range(len(confirmed))], 
               new_order)
    
    #plt.grid(b=True, which='minor', color='#666666', linestyle='-')
    #plt.grid()
    plt.legend()
    plt.minorticks_on()
    #plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
    plt.rcParams['axes.facecolor'] = 'whitesmoke'
    plt.savefig('increase_over_months')
```

```python
display_barplot(df_ww2,['Confirmed','Recovered','Deaths'])
```

## Data Modeling Part 2 -  Covid Numbers (distribution across ages - countries) 
#### age distribution

```python
#loading the data
df_v2= pd.read_csv('./data/COVID19.csv')
```

```python
df_v2 = df_v2[['ID','age','sex','country']]
```

```python
len(df_v2.query('age.isnull()')),len(df_v2.query('age.notna()'))
```

## Prepare Data

```python
#drop nulls from age column 
covid_ages = df_v2.dropna(subset=['age'])
```

```python
# confirmation whether it worked
covid_ages.shape
```

```python
# check data type 
covid_ages['age'].dtype
```

```python
# remove-non-numeric-rows-in-one-column
# https://stackoverflow.com/questions/33961028/remove-non-numeric-rows-in-one-column-with-pandas
covid_ages = covid_ages[covid_ages['age'].str.isnumeric()]
```

```python
# pandas-convert-dtype-object-to-int
# https://stackoverflow.com/questions/39173813/pandas-convert-dtype-object-to-int
covid_ages['age'] = covid_ages['age'].astype(str).astype(int)
```

```python
#https://stackoverflow.com/questions/49382207/how-to-map-numeric-data-into-categories-bins-in-pandas-dataframe
bins = [0, 15, 30, 45, 60, 75, 90, np.inf]
names =['<15','15-30','30-45','45-60','60-75','75-90','>90']
covid_ages['AgeRange'] = pd.cut(covid_ages['age'], bins, labels=names)
```

```python
covid_ages.head(1)
```

```python
base_color =sns.color_palette()[0]
age_counts=covid_ages.AgeRange.value_counts()
total=covid_ages.shape[0]
```

```python
plt.figure(figsize=[9,6])
sns.countplot(data=covid_ages, y='AgeRange',color=base_color)
plt.title('Distribution of Dataset Across Age Groups')
plt.xlabel('number of confirmed diseases')
for i in range(age_counts.shape[0]):
    count=age_counts[i]
    pct_string = '{:0.1f}%'.format(100*count/total)
    plt.text(count+1, i , pct_string, va = 'center')
plt.savefig('distribution_age')
```

#### country distribution (top 5)

```python
# countries-aggregated data w
df_v3 = pd.read_csv('./data/countries-aggregated_csv.csv', parse_dates=True)
```

```python
covid_top5 = df_v3.groupby('Country').Confirmed.sum().reset_index()
covid_top5 = covid_top5.sort_values(by=['Confirmed'], ascending=False)   
covid_top5.head()
```

```python
top5_countries = list(covid_top5.Country.head(5).values)
top5_countries 
```

```python
df_v3['Date'] = pd.to_datetime(df_v3['Date'])
```

```python
pd.options.mode.chained_assignment = None
covid_countries = df_v3.query('Date >= "2020-03-01"')
# generate name of month of from date
# https://stackoverflow.com/questions/57033657/how-to-extract-month-name-and-year-from-date-column-of-dataframe/57033774
covid_countries['Month'] = covid_countries['Date'].dt.strftime('%b')
covid_countries.head(3)
```

```python
covid_countries = covid_countries.groupby(['Country','Month']).Confirmed.sum().reset_index()
covid_countries.head()
```

```python
#https://stackoverflow.com/questions/33990955/combine-pandas-dataframe-query-method-with-isin
covid_countries=covid_countries.query('Country in @top5_countries')
```

```python
covid_countries.head()
```

```python
countries = covid_countries.pivot("Country", "Month", "Confirmed")
countries
```

```python
#https://stackoverflow.com/questions/36346071/pandas-pivot-table-changing-order-of-non-index-columns
column_order = ['Mar', 'Apr', 'May','Jun']
row_order = top5_countries 
countries = countries.reindex(column_order, axis=1)
countries = countries.reindex(row_order, axis=0)
```

```python
countries
```

```python
sns.set(rc={'figure.figsize':(7, 5)})
ax = sns.heatmap(countries, annot=True, fmt="d",cmap="YlGnBu")
plt.savefig('heatmap_top5')
```
